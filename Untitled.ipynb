```python
def label_diff(i,j,text,X,Y, up = 0):
    x = (X[i]+X[j])/2
    y = max(Y[i], Y[j])
    dx = abs(X[i]-X[j])

    props = {'connectionstyle':'bar','arrowstyle':'-',\
                 'shrinkA':15,'shrinkB':15,'linewidth':1}
    ax.annotate(text, xy=(X[i]+0.5,y+up), zorder=0, fontsize = 12)
    ax.annotate('', xy=(X[i]+0.1,y), xytext=(X[j]-0.1,y), arrowprops=props)
    
    
saHBPmP = saHBP.mean(axis = 1)
smHBPmP = smHBP.mean(axis = 1)
saLBPmP = saLBP.mean(axis = 1)
smLBPmP = smLBP.mean(axis = 1)
fig2, ax = plt.subplots(figsize = (5,6), dpi = 1000)
p = ax.boxplot([saHBPmP, smHBPmP, saLBPmP, smLBPmP], labels = ['NHBP', 'OHBP','NLBP', 'OLBP'], manage_ticks  = True)


label_diff(0,1,'ns', [1,2], [17,18], up = 1.5)
label_diff(0,1,'ns', [3,4], [17, 15], up = 1.5)
label_diff(0,1,'* * *', [1,3], [24, 25], up = 2.5)
label_diff(0,1,'* * *', [2,4], [24, 26], up = 2.5)

ax.set_ylim(9,30)
ax.set_ylabel(r'BP $(\frac{\mu g}{L})$')
ax.set_title('Mean BP values distribution')
fig2.savefig('uno.png')
```

```python
def RR(incE, incNE, n1):
    rr = (incE / n1) / (incNE / n1)
    a = ((n1 - incE) / incE) / n1
    b = ((n1 - incNE) / incNE) / n1
    ich = np.log(rr) + (1.96 * np.sqrt((a+b)))
    icl = np.log(rr) - (1.96 * np.sqrt((a+b)))
    return rr, [np.exp(icl), np.exp(ich)]
```

```python
np.random.seed(12)
def EvenOcN(center, sigma, inc):
    ranor = np.random.normal(center, sigma, 200)
    events = inc.count(1)
    r = []
    while len(r) < events:
        x = list(np.random.choice(ranor, 1))
        y = [int(i) for i in x if int(i) % 2 == 0 and int(i) <=30 and int(i>0)]
        r += y
    return r


saHBPtoc = EvenOcN(15, 3, saHBPinc)
smHBPtoc = EvenOcN(12, 3, smHBPinc)
saLBPtoc = EvenOcN(24, 3, saLBPinc)
smLBPtoc = EvenOcN(20, 3, smLBPinc)
```

```python
np.random.seed(40)
def biasData(ev, psex, page, sm):
    nsex = int(ev * psex)
    nage = int(ev * page)
    w = nsex * [1]
    h = (ev - len(w)) * [0]
    sexh = list(np.random.randint(0, 2, saHighN - ev))
    
    oldnorm = np.random.normal(28, 5, 400)
    youngnorm = np.random.normal(26, 3, 400)
    restnorm = np.random.normal(28, 3, 400)
    old = []
    young = []
    rest = []
    while len(old) < nage:
        o = np.random.choice(oldnorm, 1)
        if int(o) < 42 and int(o) > 32:
            old.append(int(o))
    while len(young) < (ev-nage):
        y = np.random.choice(youngnorm, 1)
        if int(y) > 20 and int(y) < 32:
            young.append(int(y))
    while len(rest) < (saHighN - ev):
        r = np.random.choice(restnorm, 1)
        if 20<int(r)<42:
            rest.append(int(r))
    #ageh = list(np.random.randint(20,42, saHighN - ev))
    status = [0] * saHighN
    if sm:
        status = [1] * saHighN
    
    return sexh + w + h, rest + young + old, status


saHBPsex, saHBPage, saHBPstat = biasData(saHBPev, 0.75, 0.65, False)
smHBPsex, smHBPage, smHBPstat = biasData(smHBPev, 0.75, 0.8, True)
saLBPsex, saLBPsex, saLBPstat = biasData(saLBPev, 0.5, 0.5, False)
smLBPsex, snLBPsex, smLBPstat = biasData(smLBPev, 0.5, 0.65, True)
```
